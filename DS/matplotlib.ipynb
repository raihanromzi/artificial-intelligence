```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 10)
x
```

```python
y = x ** 2
y
```

```python
# Functional
plt.plot(x, y)
plt.title('Plotting x and y')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(2, 2, 2)
plt.plot(y, x, 'b')

plt.subplot(2, 2, 3)
plt.plot(x, y, 'g')

plt.subplot(2, 2, 4)
plt.plot(y, x, 'y')

plt.show()
```

```python
# OOP methiods
fig = plt.figure() # Imaginary canvas/figure
axes = fig.add_axes([0.3, 0.1, 1.0, 0.9]) # x, y, width, height
axes.plot(x, y, 'r')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Plotting x and y')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.0, 2.0, 0.8, 0.8])
axes2 = fig.add_axes([1.0, 2.0, 0.8, 0.8])
axes3 = fig.add_axes([0.0, 1.0, 0.8, 0.8])
axes4 = fig.add_axes([0.0, 0.0, 0.8, 0.8])
axes5 = fig.add_axes([-1.0, 0.0, 0.8, 0.8])

axes1.set_title('axes1')
axes1.plot(x, y, 'r')

axes2.set_title('axes2')
axes2.plot(y, x, 'b')

axes3.set_title('axes3')
axes3.plot(x, y, 'g')

axes4.set_title('axes4')
axes4.plot(y, x, 'y')

axes5.set_title('axes5')
axes5.plot(y, x, 'k')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y, 'r')
```

```python
# Subplot in a figure
# Membuat add_axes secara otomatis berdarkan nrows, ncols
fig, axes = plt.subplots(nrows=1, ncols=2) # Same as fig = plt.figure()
# axes.plot(x, y, 'r')

axes[0].plot(x, y, 'r')
axes[0].set_title('axes[0]')
axes[1].plot(y, x, 'b')
axes[1].set_title('axes[1]')


# # Bisa nge loop di axes
# for ax in axes:
#     ax.plot(x, y, 'r')


# Merapihkan overlapping plot, at the end of the code
plt.tight_layout()
```

```python
# Figure size and DPI
fig = plt.figure(figsize=(4, 2))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x, y, 'r')
```

```python
fig, axes = plt.subplots(figsize=(8, 2), nrows=1, ncols=2)
axes[0].plot(x, y, 'r')
axes[1].plot(y, x, 'b')

plt.tight_layout()
```

```python
# Save figure to file
fig.savefig('plot.png', dpi=100)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(x, y, 'r', label='X label')
ax.plot(y, x, 'b', label='Y label')
ax.set_title('Plotting x and y')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')

# Legend
# ax.legend(loc=0)
ax.legend(loc=(0.6, 0.6))
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 4))
ax.plot(x, y, label='X label', color='purple', lw=3, ls='--', marker='o', markerfacecolor='red', markersize=12, alpha=1,
        markeredgecolor='black', markeredgewidth=4)
ax.set_xlim([0, 2])
ax.set_ylim([0, 10])
ax.legend(loc=0)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 0.8, 0.8])
ax.plot(x, y, color='purple', linewidth=4, linestyle='-', alpha=1, marker='o', markersize=20, 
        markerfacecolor='yellow', markeredgewidth=4, markeredgecolor='red')
ax.set_xlabel('X Label')
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(8, 6))
ax[0].scatter(x, y, label='X label', color='purple', s=100, marker='o', edgecolor='red', linewidth=4)
ax[1].hist(x, bins=10, color='purple', edgecolor='red', linewidth=4)
fig.tight_layout()
```
